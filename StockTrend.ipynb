# Model and Numerical Methods


In this project we develpped a stock analyzer that uses historical price data from the Yahoo Finance Library (yfinance). In this report we modeled the relationship between the time (days, months, years) and the stocking closing price using a linear regression. Although this report will be demonstrating the result for 3 major stocks over 3 different time periods, our code is generalized for any stock available on the market as long as its ticker is valid on Yahoo Finance. So the goal is to essentially create and fit a line that best fit throughout data points and observe if a real trend exists; and if it does it would be able to inform us on trading decision (Buy or Sell).

As previously mentioned, we modeld the relationship between time (as the index of trading days) and the stock's daily closing price using linear regression, which is a nice technique for prediction and data analysis.

```python
#IMPORTS
import yfinance as yf
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

welcome = print ("Welcome to the stock analyzer, where we use both the slope direction and the R squared value to inform you to buy or sell your favorite stock!")
def tickers(inp):
    if inp == "help":
        print ("Here are some common companies to analyze! \n AAPL (Apple), GOOGL (Alphabet), MSFT (Microsoft), AMZN (Amazon), META (Meta), TSLA (Tesla), GS (Goldman Sachs), DJIA (Dow Jones Industrial Average)")
    else:
        return None
naming = input ("If you'd like a list of common tickers please type (help). If you already know the ticker you want to use, type (no): ")

tickers(naming)

stock = input ("Please enter the tag of the stock to analyze: ") # Code to allow the user to input the TAG of the stock they would like to analyze

#STOCK and PERIOD only able to change those 2 parameters
period = input ("Please enter the period you would like to analyze, for long term investors, it's recommended to use a period of 5 years to base your decision. Your options are (5d, 1mo, 6mo, 1y, 2y, 5y): ")   #Only possible options: "1d", "5d", "1mo", "6mo", "1y", "2y", "5y"


data = yf.download(stock, period = period, interval="1d") 
prices = data["Close"] 
dates = data.index 

#Check if any data was returned
if prices.empty:
    print("No data found. Check stock symbol or internet connection.")
else:
    #prepare the data for linear regression
    x = (dates - dates[0]).days.values.reshape(-1, 1)
    y = prices.values.reshape(-1, 1)

    #FIT the linear regression model
    model = LinearRegression()
    model.fit(x, y)
    slope = model.coef_[0][0]     
    r_squared = model.score(x, y)
    predicted_y_value = model.predict(x)

 
    start_price = float(prices.iloc[0])   
    end_price = float(prices.iloc[-1])
    percent_change = ((end_price - start_price) / start_price) * 100
    max_price = float(prices.max())
    min_price = float(prices.min())
    trend = "Upward " if slope > 0 else "Downward "
    answer_based_r2 = "The R^2 term is greater than 0.7 therefore the coefficient of determination is STRONG" if r_squared > 0.7 else "The R^2 term is greater than 0.7 therefore the coefficient of determination is WEAK"
    buy_or_sell = "Based on the slope AND the R squared value, we should BUY the stock" if slope > 0 and r_squared >0.7  else "SELL the stock as the slope is trending downwards OR the R squared value is too low (weak)"
                 
    print(f"Analysis for       {stock} over {period} ")
    print(f"Starting Price:    {start_price:.2f}$")
    print(f"End Price:         {end_price:.2f}$")
    print(f"Change over time:  {end_price - start_price:.2f}$ ({percent_change:.2f}%)")
    print(f"Highest Price:     {max_price:.2f}$")
    print(f"Lowest Price:      {min_price:.2f}$")
    print(f"Trend Slope:       {slope:.4f} USD")
    print(f"R^2 Value:         {r_squared:.4f}")
    print(f"Trend Direction:   {trend}")
    print(f"Buy/Sell decision: {buy_or_sell}")

    print("------------------------------------------\n")
    
    #Prediction for future investment
    amount_invested = float(input("Next we will predict how much money you will gain/lose. How much money are you investing? : "))

    # Define years and convert them to number of days
    predict_years = [1, 2, 3, 10, 20]
    predict_days = [365 * y for y in predict_years]

    returns = []
    predicted_prices = []

    # Calculate future price and investment value
    for days in predict_days:
        predicted_price = max(0, start_price + slope * days)
        shares = amount_invested / start_price
        future_value = shares * predicted_price
        predicted_prices.append(predicted_price)
        returns.append(max(0, future_value)) 

    # Create and print the DataFrame
    df = pd.DataFrame({
        "Years in Future": predict_years,
        "Predicted Stock Price ($)": [f"{p:.2f}" for p in predicted_prices],
        "Estimated Investment Value ($)": [f"{v:.2f}" for v in returns]
    })

    print("\nEstimated Future Long Term Investment Value:")
    print(df)


    print("------------------------------------------\n")
    #Ploting the graph
    plt.figure(figsize=(16, 8)) 
    plt.plot(dates, prices, label="Actual Prices", linewidth=2)
    plt.plot(dates, predicted_y_value, label="Regression Line", linestyle="--", linewidth=2)
    plt.title(f"{stock} Stock Price with Linear Regression ({period}). \n {answer_based_r2}", fontsize=16)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Price (USD)", fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```
